```python
corpus = [
    "low",
    "lower",
    "lowest",
    "new",
    "newer",
    "widest"
]

print(corpus)
```

```python
corpus = [word.lower() for word in corpus]
print("Preprocessed Corpus:")
print(corpus)
```

```python
vocab = {}
for word in corpus:
    chars = " ".join(list(word)) + " </w>"
    vocab[chars] = vocab.get(chars, 0) + 1
print("Character-Level Vocabulary:\n")
for word, freq in vocab.items():
    print(word, ":", freq)
```

```python
print("Word Frequencies:\n")
for word, freq in vocab.items():
    print(f"{word} --> {freq}")
```

```python
def get_pair_counts(vocab):
    pair_counts = {}
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols) - 1):
            pair = (symbols[i], symbols[i + 1])
            pair_counts[pair] = pair_counts.get(pair, 0) + freq
            return pair_counts
```

```python
pairs = get_pair_counts(vocab)
print("Adjacent Pair Frequencies:\n")
for pair, count in pairs.items():
    print(pair, "-->", count)
```

```python
best_pair = max(pairs, key=pairs.get)
print("Most Frequent Pair:")
print(best_pair)
print("Frequency:", pairs[best_pair])
```

```python
def merge_pair(pair, vocab):
    new_vocab = {}
    bigram = " ".join(pair)
    replacement = "".join(pair)
    for word in vocab:
        new_word = word.replace(bigram, replacement)
        new_vocab[new_word] = vocab[word]
        return new_vocab
```

```python
vocab = merge_pair(best_pair, vocab)
print("Updated Vocabulary:\n")
for word, freq in vocab.items():
    print(word, "-->", freq)
```

```python
num_merges = 10
merge_rules = []
for i in range(num_merges):
    pairs = get_pair_counts(vocab)
    if len(pairs) == 0:
        break
        best_pair = max(pairs, key=pairs.get)
        merge_rules.append(best_pair)
        print(f"\nMerge {i+1}: {best_pair}")
        vocab = merge_pair(best_pair, vocab)
```

```python
print("Learned Merge Rules:\n")
for i, rule in enumerate(merge_rules, start=1):
    print(f"{i}. {rule}")
```

```python
final_vocab = set()
for word in vocab:
    tokens = word.split()
    for token in tokens:
        final_vocab.add(token)
        print("Final Vocabulary:\n")
        print(sorted(final_vocab))
        print("\nVocabulary Size:", len(final_vocab))
```
