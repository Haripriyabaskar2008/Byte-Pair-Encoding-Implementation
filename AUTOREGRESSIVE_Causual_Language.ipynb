```python
import numpy as np
```

```python

with open("corpus.txt","r") as file:
    corpus = file.read().splitlines()

print("Corpus:")
print(corpus)
```

```python
corpus = [sentence.lower().strip() for sentence in corpus]

print("Preprocessed Corpus:")
print(corpus)
```

```python
tokenized = []

for sentence in corpus:
    words = sentence.split()
    tokenized.append(words)

print("Tokenized Corpus:")
print(tokenized)
```

```python
vocab = {}
index = 0

for sentence in tokenized:

    for word in sentence:

        if word not in vocab:

            vocab[word] = index
            index += 1

print("Vocabulary:")
print(vocab)
```

```python
token_ids = []

for sentence in tokenized:

    ids = [vocab[word] for word in sentence]

    token_ids.append(ids)

print("Token IDs:")
print(token_ids)
```

```python
input_sequences = []
target_sequences = []

for ids in token_ids:
    input_sequences.append(ids[:-1])
    target_sequences.append(ids[1:])

print("Input Sequences:")
print(input_sequences)

print("\nTarget Sequences:")
print(target_sequences)
```

```python
embedding_dim = 4

embedding_matrix = np.random.rand(len(vocab), embedding_dim)

print("Embedding Matrix:\n")
print(embedding_matrix)
```

```python
embedded_sequences = []

for sequence in input_sequences:
    embedded = embedding_matrix[sequence]
    embedded_sequences.append(embedded)

print("Embedded Input:\n")
print(embedded_sequences[0])
```

```python
position = np.arange(len(input_sequences[0])).reshape(-1,1)

embedded_sequences[0] = embedded_sequences[0] + position

print("After Positional Encoding:\n")
print(embedded_sequences[0])
```

```python
sequence_length = len(input_sequences[0])

causal_mask = np.tril(np.ones((sequence_length, sequence_length)))

print("Causal Mask:\n")
print(causal_mask)
```

```python
Q = embedded_sequences[0]
K = embedded_sequences[0]
V = embedded_sequences[0]

attention_scores = np.dot(Q, K.T)

print("Attention Scores:\n")
print(attention_scores)
```

```python
masked_scores = attention_scores * causal_mask

print("Masked Attention Scores:\n")
print(masked_scores)
```

```python
exp_scores = np.exp(masked_scores)

softmax = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)

print("Softmax Output:\n")
print(softmax)
```

```python
attention_output = np.dot(softmax, V)

print("Attention Output:\n")
print(attention_output)
```

```python
output_weights = np.random.rand(embedding_dim, len(vocab))

logits = np.dot(attention_output, output_weights)

print("Logits:")
print(logits)
```

```python
predicted_token = np.argmax(logits, axis=1)

print("Predicted Token IDs:")
print(predicted_token)
```

```python
probabilities = np.exp(logits)
probabilities = probabilities / np.sum(probabilities, axis=1, keepdims=True)

loss = 0

for i in range(len(target_sequences[0])):
    loss -= np.log(probabilities[i][target_sequences[0][i]])

loss = loss / len(target_sequences[0])

print("Cross Entropy Loss:", loss)
```

```python
id_to_token = {value: key for key, value in vocab.items()}

print("Generated Tokens:\n")

for token in predicted_token:
    print(id_to_token[token])
```
